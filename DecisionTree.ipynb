```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
col_name = dataset.feature_names
```

```python
x = dataset.data #input value
```

```python
t= dataset.target #target value
```

```python
np.unique(t) # remove duplicate
```

```python
x.shape
```

```python
t.shape
```

```python
x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.3, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
t_train.shape
```

```python
t_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state=0)
```

```python
model.fit(x_train, t_train)
```

```python
print(f'train score: {model.score(x_train, t_train)}')
print(f'test score: {model.score(x_test, t_test)}')     #97.8% accurate
```

```python
model.predict(x_test)
```

```python
feature_importance = model.feature_importances_ # petal width is most important and sepal length is least important
```

```python
y = col_name
width = feature_importance
plt.barh(y, width)
plt.show()
```
